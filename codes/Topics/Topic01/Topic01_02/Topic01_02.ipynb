# Topic 1.2 - Development of Business Analytics

## 1. Background of Business Analytics Evolution

### (1) From HIPPO to data-driven decision-making

Traditional HIPPO (Highest Paid Person’s Opinion) decision-making approaches are no longer sufficient.

Modern business environments are faster and more complex.

Businesses now generate large volumes of **data** from many sources:

- For example:

    - Customer transactions
    - Social media activity
    - Financial market data
    - Supply chain records

- Even a single online purchase produces multiple data points, such as:

    - What the customer viewed 
    - How long they hesitated 
    - Which payment method they chose 
    - How they arrived at the site

The key challenge is not data collection, but knowing **how to use the data effectively.**

### (2) Challenges Addressed by Business Analytics

Business analytics uses data and analytical tools to support better decision-making, not only build models or dashboards.

- It combines statistical thinking, technological capability, and business insight.
- At its core, business analytics helps organizations answer a fundamental strategic question: **What should we do next?**

Business analytics focuses on turning data into insight and action

- Analysts help organizations interpret what data reveals about: for example:

    - Customer needs
    - Risks and regulations
    - Investor sentiment
    - Competitive dynamics
    - Operational bottlenecks

- These insights are translated into actionable decisions

Business analytics improves how decisions are made, while decision-making fundamentals remain unchanged:

- Decision-making still involves uncertainty and trade-offs
- Business analytics enhances this process by:

    - Using data as guidance
    - Applying structured analytical methods
    - Moving from educated guesswork to informed action

## (2) Stages of Evolution of Business Analytics

The development of business analytics has gone through four stages, reflecting the shift from intuition-based decision-making to data-driven and evidence-based approaches.

### (1) Stage 1 - Gut Instinct (The HIPPO Era)

Stage Background: For most of the 20th century, decisions were dominated by experience, instinct, and authority of leadership.

Stage Characteristics:

- HIPPO (Highest Paid Person’s Opinion) Model: The most senior person in the room often had the final say - regardless of what the data might suggest.

- This approach often lacks transparency and makes it difficult to learn from both successes and failures, while experience remains valuable.

### (2) Stage 2 - **Simple Data Support**

Stage Background: With the rise of digital systems, businesses began generating and storing data and using it to create basic reports and forecasts.

Stage Characteristics:

- Data was mainly **descriptive**: it helped explain what had happened, e.g., monthly sales reports or annual budget reviews.

- Dashboards and spreadsheets became common tools for tracking performance.

- But, data was not used to predict what to do next.

### (3) Stage 3 - **Advanced Analytics**

Stage Background: As computing power increased, so did analytical ambition.

Stage Characteristics:

- Businesses began using **predictive models** and machine learning to forecast future outcomes and identify hidden patterns.

- This allowed for more proactive decisions:

    - Anticipating which customers might leave

    - Deciding when to issue stock to raise capital

    - Optimizing prices dynamically.

- However, a model is valuable only if it leads to better decision-making.

### (4) Stage 4 - **Experimentation and Causal Learning**

Stage Background: The frontier today is not just prediction - it’s learning what actually works.

Stage Characteristics:

- Companies now use prescriptive methods to evaluate strategies and understand the true impact of different choices:

    - A/B testing

    - Field experiments

    - Causal inference

- This lets firms go beyond correlation and ask the crucial question: What happens if we actually do X instead of Y?

## 3. Implications of Evolution of Business Analytics

Business decision-making has gone through a quiet revolution:

- The tools, habits, and expectations surrounding decision-making have shifted dramatically.

- Business analytics sits at the center of this transformation.

- However, even as decision-making methods evolve, the core problems addressed by business analytics remain unchanged, such as:

    - Who should we hire?

    - Where should we invest?

    - How much should we spend on advertising?
