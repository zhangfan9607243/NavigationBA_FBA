# Topic 1.3 - The Roles of Business Analyst

## 1. The Idea of Business Analyst

### (1) Definition of Business Analyst

A Business Scientist is someone who help organisations make better decisions by:

- Combining data

- Applying domain knowledge

- Using structured reasoning

The term emphasises:

- A blend of analytical skills

- Conceptual thinking

- Real-world business impact

It reflects not only what analysts do, but how they think and the value they bring to organisations.

### (2) Fundamental Skills of Business Analyst

Business analysts need a blend of skills:

- Business understanding:

    - Business analysts must understand the context

    - For example: What is the decision? What matters to stakeholders? What constraints shape the options?

- Data analysis:

    - Business analysts need to clean, explore, and model data using the right tools

    - This includes descriptive, predictive and causal tools to support strategy

- Communication and storytelling:

    - Business analysts must craft narratives that are statistically sound

    - These narratives must also be understandable, relevant, and actionable

These skills grow through practice:

- Working on real problems

- Engaging with real data

- Reflecting on how analysis shapes action

These skills are not tied to a specific tool or business discipline, but influence how analysts approach problems and turn analysis into action.

### (3) Business Thinking

#### (a) Mindset of Business Analyst

Business analytics is more than learning tools

- Many students expect business analytics to focus mainly on technical skills, such as:

    - Writing code

    - Creating charts and visualisations

    - Building analytical models

- However, the foundation of business analytics goes beyond technical abilities.

- It involves:

    - Using data to solve complex and ambiguous business problems

    - Developing judgement, not just technical skills

    - Knowing how to ask better questions and select the appropriate analytical tools

Business scientists must think about the problem conceptually before working with data.

- This includes:

    - Understanding the business context

    - Identifying what matters most

    - Recognising different types of analytical questions

- Types of questions that shape analysis:

    - Descriptive questions – What happened?

    - Predictive questions – What might happen?

    - Causal questions – Why did it happen?

- Analysis should therefore be treated as a problem-solving process, not just a technical task.

#### (b) Thinking Like a Business Scientist

Science is about systematically learning from the world:

- Scientists typically follow a structured process:

    - Start with questions

    - Develop hypotheses

    - Design methods to test those hypotheses

    - Use evidence to update their beliefs

- Key principles of scientific thinking:

    - Accept uncertainty

    - Ask clear and precise questions

    - Let data guide conclusions

    - Stay open to the possibility of being wrong

Business scientists apply the same scientific mindset in a commercial environment.

- Although, the tools may be different and decisions often need to be made more quickly

- The mindset remains the same:

    - Stay curious

    - Stay structured

    - Stay open to what the data reveals

Business scientists often investigate practical business problems: 

- For example:

    - Why a product is not selling as expected

    - Whether a loyalty program increased customer retention

    - How to segment customers more effectively

- These questions are important because business decisions and outcomes depend on them

Business scientists do more than simply report numbers. 

- Business scientists:

    - Form hypotheses

    - Test ideas using available data

    - Work with messy constraints and incomplete information

    - Analyse data under real-world pressure

- Their goal is to:

    - Build a clear evidence-based explanation

    - Help stakeholders (e.g., managers, clients, teams) make better decisions

#### (c) From Mindset to Practice

The business scientist’s mindset guides how problems are approached in practice:

- Key idea is that:

    - There is no fixed recipe or step-by-step formula

    - Instead, analysts follow a structured way of thinking throughout the analysis process

- This thinking guides every stage of data analysis and decision-making

**Key steps in the business scientist approach**:

1. **Forming hypotheses**

    - Before analysing data, a business scientist asks:

        - What factors might be driving this outcome?

        - If this explanation were true, what patterns should appear in the data?

    - Important principle:

        - Do not jump into data blindly

        - Start with possible explanations

2. **Testing ideas with data**

    - Data is treated as evidence:

        - An idea is not valid just because you believe it

        - It must be supported by data

    - Evidence is used to evaluate whether hypotheses hold up.

3. **Iterating and refining**

    - Analytics rarely works perfectly on the first attempt.

    - The process often involves:

        - Exploring data

        - Learning from results

        - Revising assumptions

        - Improving the analysis

    - Iteration is a normal and necessary part of good analytics.

4. **Communicating clearly**

    - Business scientists must translate complex analysis into clear insights.

    - This involves:

        - Presenting findings in an understandable way

        - Explaining what the results mean

        - Telling a story with data that helps others take action

The **three roles of a business scientist** under this mindset are:

1. **Detective**

    - Investigates data

    - Identifies patterns and hidden insights

    - Tests possible explanations

2. **Strategist**

    - Connects analysis to business goals

    - Interprets results for decision-making

    - Guides business strategy

3. **Storyteller**

    - Communicates insights clearly

    - Turns data into meaningful narratives

    - Helps stakeholders understand and act on findings

## 1. The Interpreter Role of Business Analyst

Business analysts requires decisions and skills in interpretation, judgment, and communication - more than just numbers, algorithms, or datasets:

- Business analysts sit at the intersection of data and decision-making.

    - They are not just technicians or model-builders.

    - They are **interpreters** to make data meaningful to the people who need to act on it.

- Business analyst needs a blend of skills: business understanding, analytical fluency, and communication.

    - This skillset is not limited to advanced degrees - undergraduate analysts with proper training can create significant value from the start.

    - The course is designed to build this foundational skillset for effective business analysts.

> Citation:
> "MBAs who have good business intuition but also speak the language of statisticians are rock stars."
> -- Susan Athey, Professor of Economics at Stanford

Business analysts are modern storytellers:

- Business analysts use data to tell clear and truthful stories that support better decision-making.

- Business analyst’s role is to build a bridge between the data and the decision, for example:

    - Why sales are falling

    - How investor sentiment is improving

    - What might happen if the firm raises prices or changes strategy



## 3. Collaboration Role of Business Analysts

Business analysts have to work with:

-   Data engineers - build the infrastructure and pipelines that make analysis possible.

-   Business stakeholders - define the problems and act on the results.

-   Other analysts - bring complementary expertise and perspectives.

Business analysts are glue in the collaboration - keeps the analytical process connected to business goals, ethical standards, and practical decisions.
